```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
plt.show()
```

```python
# First Method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
plt.plot();
```

```python
# Second Method;
fig = plt.figure()
ax = fig.add_axes([1,2,3,4])
ax.plot(x, y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x, y)
plt.show()
```

```python
# Third Method (Recommended *)
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
fig, ax = plt.subplots()
plt.plot(x,y);
```

```python
fig, ax = plt.subplots()
x = [1,2,3,4,5]
y = [11,22,44,66,999]
plt.plot(x,y);
type(fig), type(ax)  # OOP paradise
```

```python
# My Way :)  *
fig, ax = plt.subplots()
ax.plot([1,2,3],[33,369,999]);
```
